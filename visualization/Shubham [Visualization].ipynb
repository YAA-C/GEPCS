## Q1) Which Guns are used more? Show uses in form of piechart / barchart. Uses are indicated when a shot is registered i.e. weapon fired and shot hit target, weaponUsed feature is provided for weapon used. 

```python
import pandas as pd
import matplotlib.pyplot as plt

dfWeapon = pd.read_csv("../DemoFiles/csv/1win-vs-mibr-m1-vertigo-245820.csv")

# Group by "weaponUsed" and calculate the sum of "dmgDone" for each weapon
weapon_dmg_sum = dfWeapon.groupby("weaponUsed")["dmgDone"].sum()

# Sort the values in descending order
weapon_dmg_sum = weapon_dmg_sum.sort_values(ascending=False)

# Set a larger figure size
plt.figure(figsize=(10, 6))  # Adjust the width and height as needed

# Create a bar chart
plt.bar(weapon_dmg_sum.index, weapon_dmg_sum.values)

# Add labels and title
plt.xlabel("Weapon Used")
plt.ylabel("Total Damage Done")
plt.title("Total Damage Done by Each Weapon")

# Rotate x-axis labels and remove 'weapon_' prefix
labels = [label.replace('weapon_', '') for label in weapon_dmg_sum.index]
plt.xticks(ticks=range(len(labels)), labels=labels, rotation=45)  # You can adjust the rotation angle as needed

# Show the plot
plt.show()
```

## Q2) Which Body part is mostly targetted by players? Show in form of piechart / barchart. When a shot is registered i.e. weapon fired and shot hit target targetHitArea feature is given to show where shot was landed.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("../DemoFiles/csv/1win-vs-mibr-m1-vertigo-245820.csv")
df

hit_area_mapping = {
    0.0: 'GENERIC',
    1.0: 'HEAD',
    2.0: 'CHEST',
    3.0: 'STOMACH',
    4.0: 'LEFTARM',
    5.0: 'RIGHTARM',
    6.0: 'LEFTLEG',
    7.0: 'RIGHTLEG',
    8.0: 'GEAR',
    'nan': 'NaN'
}

# Map the values in the DataFrame to the corresponding categories
df['targetHitArea'] = df['targetHitArea'].map(hit_area_mapping)

# Set a stylish Seaborn theme
sns.set_theme()

# Create a bar chart
plt.figure(figsize=(10, 6))
sns.countplot(x='targetHitArea', data=df, hue='targetHitArea', palette='viridis', dodge=False)
plt.xlabel('Target Hit Area')
plt.ylabel('Count')
plt.title('Bar Chart of Target Hit Area')
plt.tight_layout()  # Adjust layout for better spacing
plt.legend().set_visible(False)  # Hide the legend
plt.show()
```

## Q3) Which Guns Category are used more? Show uses in form of piechart / barchart. Uses are when a shot is registered i.e. weapon fired and shot hit target, usefull feature is weaponCategory.

```python
import pandas as pd
import matplotlib.pyplot as plt

dfCat = pd.read_csv("../DemoFiles/csv/auto0-20231225-155107-1828335508-de_dust2-_cz__siddhesh_game_server-640472.csv")

# Group by "weaponCategory" and calculate the sum of "dmgDone" for each category
category_dmg_sum = dfCat.groupby("weaponCategory")["dmgDone"].sum()

# Sort the values in descending order
category_dmg_sum = category_dmg_sum.sort_values(ascending=False)

# Remove 'weapon_category_' prefix from index for labels
labels = [label.replace('weapon_category_', '') for label in category_dmg_sum.index]

# Create a pie chart
plt.figure(figsize=(10, 6))  # Adjust the size as needed
plt.pie(category_dmg_sum.values, labels=labels, autopct='%1.1f%%')

# Add title
plt.title("Weapon Category Distribution based on Total Damage Done")

# Show the plot
plt.show()
```

## Q7) Where do players see vertically i.e. where do players focus there pitch at? Use all rows for feature pitch to show where players generally see. Use histogram for pitch starting from -90 to +90.

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("../DemoFiles/csv/1win-vs-mibr-m1-vertigo-245820.csv")
df

plt.figure(figsize=(18, 8))
plt.hist(df['pitch'], bins=range(-90, 100, 1), edgecolor='black')
plt.xlabel('Pitch')
plt.ylabel('Count')
plt.yscale("log")
plt.title('Histogram of Pitch')
plt.show()
```

## Q9) Whether players shoot players with scope on or scope off. Count the number of scope on/off occurrences when a shot is registered i.e. weapon fired and shot hit target. Show in piechart. Relevant features isScoping.

```python
import pandas as pd
import matplotlib.pyplot as plt

dfScope=pd.read_csv("../DemoFiles/csv/auto0-20231226-124835-1781275285-de_mirage-_cz__siddhesh_game_server-906446.csv")
dfScope

sniper_data = dfScope[dfScope['weaponCategory'] == 'weapon_category_sniper']
scoping_counts = sniper_data['isScoping'].value_counts()

# Plotting a pie chart
plt.figure(figsize=(4, 4))
plt.pie(scoping_counts, labels=scoping_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('While using Sniper does player Scopes')
plt.show()
```
