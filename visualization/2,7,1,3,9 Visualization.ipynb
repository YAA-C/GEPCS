```python
import pandas as pd
import matplotlib.pyplot as plt

dfScope=pd.read_csv("../DemoFiles/csv/1win-vs-mibr-m1-vertigo-370903.csv")
dfScope

# Count the occurrences of each unique value in the "isScoping" column
counts = dfScope["isScoping"].value_counts()

# Create a pie chart using the counts and dynamically generated labels
plt.pie(counts, labels=counts.index.astype(str), autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart of isScoping when isFiring is True")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming dfScope is your DataFrame
# Filter the DataFrame based on the condition dfScope["isFiring"] == True
filtered_df = dfScope[dfScope["isFiring"] == True]

# Count the occurrences of each unique value in the "isScoping" column
counts = filtered_df["isScoping"].value_counts()

# Create a color palette
colors = ['#66b3ff', '#99ff99', '#ffcc99']

# Define a function to format the autopct
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d}'.format(v=val) + ' (' + '{p:.2f}'.format(p=pct) + '%)'
    return my_format

# Create a pie chart using the counts and dynamically generated labels
plt.figure(figsize=(6, 4))  # Adjust the figure size
plt.pie(counts, labels=counts.index.astype(str), autopct=autopct_format(counts), startangle=90, colors=colors, wedgeprops=dict(width=0.4))
plt.title("Distribution of isScoping when isFiring is True", fontsize=16)
plt.legend(title="isScoping", loc="best", bbox_to_anchor=(0.85, 0.9, 0.5, 0.5))
plt.show()
```
