```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2) Which body part is targeted [BarChart]

```python
# 0 GENERIC	
# 1 HEAD		
# 2 CHEST		
# 3 STOMACH	
# 4 LEFTARM		
# 5 RIGHTARM	
# 6 LEFTLEG	
# 7 RIGHTLEG	
# 8 GEAR
```

```python
Unique = np.unique(df["targetHitArea"])

for i in Unique:
    print(i)
```

```python
hit_area_mapping = {
    0.0: 'GENERIC',
    1.0: 'HEAD',
    2.0: 'CHEST',
    3.0: 'STOMACH',
    4.0: 'LEFTARM',
    5.0: 'RIGHTARM',
    6.0: 'LEFTLEG',
    7.0: 'RIGHTLEG',
    8.0: 'GEAR',
    'nan': 'NaN'
}

# Map the values in the DataFrame to the corresponding categories
df['targetHitArea'] = df['targetHitArea'].map(hit_area_mapping)
```

# 2) Target Hit Area Bar Chart 

```python
# Set a stylish Seaborn theme
sns.set_theme()

# Create a bar chart
plt.figure(figsize=(10, 6))
sns.countplot(x='targetHitArea', data=df, hue='targetHitArea', palette='viridis', dodge=False)
plt.xlabel('Target Hit Area')
plt.ylabel('Count')
plt.title('Bar Chart of Target Hit Area')
plt.tight_layout()  # Adjust layout for better spacing
plt.legend().set_visible(False)  # Hide the legend
plt.show()
```

```python
count_within_range = ((df["pitch"] >= -90) & (df["pitch"] <= -10)).sum()
print("Count of entries with pitch between -90 and -10:", count_within_range)
```

```python
count_within_range = ((df["pitch"] >= 20) & (df["pitch"] <= 90)).sum()
print("Count of entries with pitch between -90 and 0:", count_within_range)
```

# 1) Most used Gun

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfWeapon=pd.read_csv("../DemoFiles/csv/auto0-20231225-155107-1828335508-de_dust2-_cz__siddhesh_game_server-640472.csv")
dfWeapon
```

```python
weapon_counts = dfWeapon["weaponUsed"].value_counts()

# Set a larger figure size
plt.figure(figsize=(10, 6))  # Adjust the width and height as needed

# Create a bar chart
plt.bar(weapon_counts.index, weapon_counts.values)

# Add labels and title
plt.xlabel("Weapon Used")
plt.ylabel("Count")
plt.title("Weapon Used Distribution")

# Rotate x-axis labels and remove 'weapon_' prefix
labels = [label.replace('weapon_', '') for label in weapon_counts.index]
plt.xticks(ticks=range(len(labels)), labels=labels, rotation=45)  # You can adjust the rotation angle as needed

# Show the plot
plt.show()
```

# 3) Gun Category used

```python
dfCat=pd.read_csv("../DemoFiles/csv/auto0-20231225-155107-1828335508-de_dust2-_cz__siddhesh_game_server-640472.csv")
dfCat
```

```python
# Get value counts
weapon_counts = dfCat["weaponCategory"].value_counts()

# Remove 'weapon_category_' prefix from index for labels
labels = [label.replace('weapon_category_', '') for label in weapon_counts.index]

# Create a pie chart
plt.figure(figsize=(10, 6))  # Adjust the size as needed
plt.pie(weapon_counts.values, labels=labels, autopct='%1.1f%%')

# Add title
plt.title("Weapon Category Distribution")

# Show the plot
plt.show()
```
